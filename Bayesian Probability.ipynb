##Load and prepare data

```python
import pandas as pd
import numpy as np

# Ensure text is lowercase for easy keyword search
df["review"] = df["review"].str.lower()
```

##Define keywords

```python
keywords = ["worst", "awful", "bad"]
```

##Compute probabilities

```python
# Total number of reviews
N = len(df)

# Determine the probability of a review being positive
p_positive = len(df[df["sentiment"] == "positive"]) / N
p_negative = 1 - p_positive

print(f"Total number of reviews: {N}")
print(f"P(Positive) = {p_positive:.4f}")
print(f"P(Negative) = {p_negative:.4f}")


import numpy as np

keywords = ["worst", "awful", "bad"]
results = []

# For each keyword
for kw in keywords:
    # Checking which reviews contain the keyword
    contains_kw = df["review"].str.contains(rf"\b{kw}\b", regex=True, case=False, na=False)


    # Marginal probability P(keyword)
    p_kw = contains_kw.mean()

    # Likelihood P(keyword|Positive)
    p_kw_given_pos = (
        df[contains_kw & (df["sentiment"] == "positive")].shape[0]
        / df[df["sentiment"] == "positive"].shape[0]
    )

    # Using Bayes' Theorem
    if p_kw > 0:
        p_pos_given_kw = (p_kw_given_pos * p_positive) / p_kw
    else:
        p_pos_given_kw = np.nan

    results.append({
        "Keyword": kw,
        "P(Positive)": round(p_positive, 4),
        "P(keyword|Positive)": round(p_kw_given_pos, 4),
        "P(keyword)": round(p_kw, 4),
        "P(Positive|keyword)": round(p_pos_given_kw, 4)
    })

results_df = pd.DataFrame(results)
results_df

```
